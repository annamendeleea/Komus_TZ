```python
# импортируем библиотеки

import pandas as pd
import datetime as dt
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from io import BytesIO
import requests


pd.set_option('display.float_format', '{:,.2f}'.format)
```

```python
#откроем таблицу №1: 

user_info = pd.read_excel('Кейс.xlsx', sheet_name='Таблица 1', skiprows=0)
user_info['Период'] = pd.to_datetime(user_info['Период'], errors='coerce')
user_info['Квартал'] = user_info['Период'].dt.quarter
user_info.head(3)
```

```python
#откроем таблицу №2: 

cohort_df = pd.read_excel('Кейс.xlsx', sheet_name='Таблица 3', skiprows=0)
cohort_df = cohort_df.fillna(0)
cohort_df.set_index('Месяц регистрации')
cohort_df['Месяц регистрации'] = pd.to_datetime(cohort_df['Месяц регистрации']).dt.date
cohort_df.head(3)
```

```python
#откроем таблицу №3: 

marketing_costs = pd.read_excel('Кейс.xlsx', sheet_name='Таблица 2', skiprows=0)
marketing_costs.columns = ['Период', 'Б', 'П', 'Д', 'Л', 'ПД', 'В']
marketing_costs['Период'] = pd.to_datetime(marketing_costs['Период'], errors='coerce')
marketing_costs['Затраты на рекламу'] = marketing_costs.sum(numeric_only=True, axis = 1)
marketing_costs['Квартал'] = marketing_costs['Период'].dt.quarter 

marketing_costs.head(3)
```

## Задача 5. 
Рассчитать LTV за каждый квартал 2021г. 

Рассчитаем LTV по формуле: LTV = (Доход - Затраты на рекламу)/Кол-во клиентов

```python
#оставляем только 2021 год:
next_year_date = '2022-01-01'
LTV_df = user_info.query('Период < @next_year_date')

#агрегируем доходы, затраты на рекламу и кол-во клиентов по кварталам
LTV_df = LTV_df.groupby('Квартал').agg({'Всего зарегистрировались': 'sum', 'Доход':'sum'})
```

```python
marketing_costs_by_quater = marketing_costs.groupby('Квартал').agg({'Затраты на рекламу': 'sum'})
LTV_df = pd.merge(LTV_df, marketing_costs_by_quater, on=('Квартал'), how ='left')


LTV_df['LTV'] = (LTV_df['Доход'] - LTV_df['Затраты на рекламу']) / LTV_df['Всего зарегистрировались']
```

## Задача 6. 

Рассчитайте ARPU за январь 2020.


В условии задачи скорее всего ошибка в дате, т.к. мы имеем данные только с января 2021 по январь 2022. 

Поэтому, будем рассчитывать ARPU за январь 2022 и сравнивать с аналогичным периодом прошлого года. 

Будем считать по формуле: ARPU = Общий доход за период / количество клиентов

```python
ARPU_df = user_info.groupby('Период').agg({'Всего зарегистрировались':'sum', 'Доход':'sum'}).reset_index()
ARPU_df['ARPU'] = ARPU_df['Доход'] / ARPU_df['Всего зарегистрировались'] 
Jan_2022 = ARPU_df.iloc[0]['ARPU']
Jan_2021 = ARPU_df.iloc[12]['ARPU']
print('январь 2022, ARPU =', float('{:.2f}'.format(Jan_2022)))
print('январь 2021, ARPU =', float('{:.2f}'.format(Jan_2021)))
diff = (Jan_2022/Jan_2021-1)*100
print('Убыль показателя, %:', float('{:.2f}'.format(diff)))
```

## Задача 7. 

Сделайте необходимые расчёты и выводы об эффективности рекламных кампаний, заполнив соответствующие пробелы в тексте


Рассчитывать окупаемость будем по формуле: 

Payback = Затраты / Прибыль


```python
Revenue = user_info.query('Период < @next_year_date')
Revenue_2021 = Revenue['Доход'].sum() 

marketing_costs_df = marketing_costs.drop(columns=['Затраты на рекламу'])
marketing_costs_df = marketing_costs_df.drop(columns=['Квартал'])
marketing_costs_df = marketing_costs_df.set_index('Период')

marketing_costs_df = marketing_costs_df.replace(to_replace=r'-\xa0', value=0, regex=True)
marketing_costs_df = marketing_costs_df.replace(to_replace='-', value=0, regex=True)
marketing_costs_df = marketing_costs_df.astype(float)


marketing_costs_2021 = marketing_costs_df.query('Период < @next_year_date')
marketing_costs_2021 = marketing_costs_2021.sum().sum()

payback_2021 = marketing_costs_2021 / (Revenue_2021-marketing_costs_2021)
print('За 2021 год команда приложения получила', Revenue_2021, 'руб. дохода от продаж,')
print('при этом расходы на рекламные кампании составили', marketing_costs_2021, 'руб.,что и обусловило отрицательные значения показателей рентабельности за данный период.')
print('Средний показатель окупаемости составляет' , float('{:.2f}'.format(payback_2021)))

```

## Задача 8. 

Заполните таблицу с показателями удержания клиентов за каждый месяц (Retention Rate) по таблице с когортами 

```python
# делим данные таблицы удержания на размеры когорт

cohort_df = cohort_df.set_index('Месяц регистрации')
retention_rate = cohort_df.div(
    cohort_df['Численность когорты'], axis=0
).drop(columns=['Численность когорты'])

print('Retention Rate')
retention_rate
```

### Конверсия

```python
#Рассчитываем конверсию
conversion = user_info.groupby('Период').agg({'Всего зарегистрировались':'sum','Купившие подписку':'sum'}).reset_index()
conversion['Конверсия'] =  conversion['Купившие подписку']/conversion['Всего зарегистрировались'] * 100

#Добавляем конверсию в датафрейм
final_table['Конверсия'] = conversion['Конверсия']
final_table
```

### ROI

ROI = ((Доходы − Затраты) / Затраты) x 100%

```python
marketing_costs_df['Total'] = marketing_costs_df.sum(axis = 1)

revenue_per_month = user_info.groupby('Период').agg({'Всего зарегистрировались':'sum','Купившие подписку':'sum', 'Доход':'sum'}).reset_index()
revenue_per_month['Доход']

data_per_month = pd.DataFrame()
data_per_month['Затраты'] = marketing_costs_df['Total']

data_per_month = pd.merge(data_per_month, revenue_per_month, on=('Период'), how ='left')
data_per_month['ROI'] = ((data_per_month['Доход'] - data_per_month['Затраты']) / data_per_month['Затраты']) * 100

data_per_month = data_per_month.set_index('Период')
ROI = data_per_month['ROI']

final_table = pd.merge(final_table, ROI, on=('Период'), how ='left')
final_table
```

### Средняя стоимость конверсии

```python
marketing_costs_df['avg_conversion_cost'] = data_per_month['Затраты'] / data_per_month['Купившие подписку']

avg_conversion_cost = pd.DataFrame()
avg_conversion_cost['Средняя стоимость конверсии'] = marketing_costs_df['avg_conversion_cost']
avg_conversion_cost

final_table = pd.merge(final_table, avg_conversion_cost, on=('Период'), how ='left')
final_table
```

```python
ARPU_df = ARPU_df.set_index('Период')
ARPU = pd.DataFrame()
ARPU = ARPU_df['ARPU']

final_table = pd.merge(final_table, ARPU, on=('Период'), how ='left')
final_table
```
